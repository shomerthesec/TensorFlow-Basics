```python
import tensorflow as tf
import csv 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_series(time, series, format="-", start=0, end=None):
            plt.plot(time[start:end], series[start:end], format)
            plt.xlabel("Time")
            plt.ylabel("Value")
            plt.grid(True)
```

```python
with open('Sunspots.csv') as csvfile :
    ds=csv.reader(csvfile, delimiter=',')
    next(ds)
    time_steps=[]
    series_data=[]
    for i in ds:
        time_steps.append(float(i[0]) )
        series_data.append(float(i[2]) )
time_steps=np.array(time_steps)
series_data=np.array(series_data)
plt.figure(figsize=(10, 6))
plot_series(time_steps, series_data)
```

```python
split_time = 3000
window_size= 20
batch_size= 32
shuffle_buffer_size= 1000
lr=3e-4
```

```python
x_train = series_data[:split_time] # the data (training) before the split_time
x_valid = series_data[split_time:] # the data (test/valid) after the split_time
time_train = time_steps[:split_time] # the time interval for the train
time_valid = time_steps[split_time:] # the time interval for the test 

def windowed_dataset(series, window_size, batch_size, shuffle_buffer):
    dataset = tf.data.Dataset.from_tensor_slices(series)
    dataset = dataset.window(window_size + 1, shift=1, drop_remainder=True)
    dataset = dataset.flat_map(lambda window: window.batch(window_size + 1))
    dataset = dataset.shuffle(shuffle_buffer).map(lambda window: (window[:-1], window[-1]))
    dataset = dataset.batch(batch_size).prefetch(1)
    return dataset

ds = windowed_dataset(x_train, window_size, batch_size, shuffle_buffer_size) 
```

```python
model=tf.keras.models.Sequential([
    tf.keras.layers.Lambda( lambda x: tf.expand_dims(x, axis=-1), input_shape=[None]),
    tf.keras.layers.Conv1D(32, 3,padding='causal', activation='relu'  ),
    tf.keras.layers.LSTM(32, return_sequences='True' ),
    tf.keras.layers.LSTM(16),
    tf.keras.layers.Dense(1),
    tf.keras.layers.Lambda(lambda x: x*100.0 )
])
```

```python
mae= tf.keras.metrics.MeanAbsoluteError()

lr_schedule = tf.keras.callbacks.LearningRateScheduler( lambda epoch: lr * 10**(epoch / window_size) ) # where lr is the lr in the optimizer

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=lr ) ,
             metrics=[mae],
             loss= 'mae')
```

```python
history=model.fit(ds, epochs= 200 , callbacks=[lr_schedule])
```

```python
tf.keras.backend.clear_session()
model=tf.keras.models.Sequential([
    tf.keras.layers.Lambda( lambda x: tf.expand_dims(x, axis=-1), input_shape=[None]),
    tf.keras.layers.Conv1D(32, 3,padding='causal', activation='relu'  ),
    tf.keras.layers.LSTM(32, return_sequences='True' ),
    tf.keras.layers.LSTM(16),
    tf.keras.layers.Dense(1),
    tf.keras.layers.Lambda(lambda x: x*100.0 )
])
```

```python

mae= tf.keras.metrics.MeanAbsoluteError()
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=lr ) ,
             metrics=[mae],
             loss= 'mae')
```

```python
history=model.fit(ds, epochs= 50 )
```

```python
forecast=[]
for time in range( len( series_data ) - window_size ): # iterating for the length of the whole data minus the windows_size as it is our step_size
    # appending the predictions of the slices of the series_data from Tn to Tn+step_size and organizing it's shape for the neural network
    # notice that the slices are on the whole series_data not just the test one
    forecast.append( model.predict( series_data[ time : time+window_size ] [np.newaxis] ) )
forecast = forecast[ split_time - window_size : ] # where the forecast is the prediction of the whole data series
#showing the output of the test data only, the unseen data output, from the split_time-window_size until the end of the data
results = np.array(forecast)[:, 0, 0] # putting the forecast data in shape for plotting
```

```python
plt.figure(figsize=(10, 6))
plot_series(time_valid, x_valid)
plot_series(time_valid, results)
```
